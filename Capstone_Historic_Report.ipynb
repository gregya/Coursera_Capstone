# Historic Site Locations

## Introduction

This project seeks to understand the location and clustering of historic sites for a given set of lattitude and longitude coordinates.

It will use the Foursquare API to find historic site data for a given location. Use cases could include: 
1. Comparing densities of historic sites within a single city for preservation  
2. Comparing cities 
3. Planning travel to various locations.

### Target Audiences
Audiences who would use this application include local city planners and researchers, as well as tourists planning visits to the area.
